```python
df.drop(columns=["fruit_subtype","fruit_label"],inplace=True)
```

```python
#splitting the x and y variable
x=df[["mass","width","height","color_score"]]
y=df["fruit_name"]
```

```python
df
```

```python
#splitting xtrain xtest y train ytest
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(xtrain)
xtrainS=scaler.transform(xtrain)
xtestS=scaler.transform(xtest)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
from sklearn.tree import  DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(xtrainS,ytrain)
predy=model.predict(xtestS)
print("Accuracy score",accuracy_score(ytest,predy),"\n")
print("classification report",classification_report(ytest,predy))
```

```python
from sklearn.ensemble import  RandomForestClassifier
model=RandomForestClassifier()
model.fit(xtrainS,ytrain)
predy=model.predict(xtestS)
print("Accuracy score",accuracy_score(ytest,predy),"\n")
print("classification report",classification_report(ytest,predy))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(xtrainS,ytrain)
predy=model.predict(xtestS)
print("Accuracy score",accuracy_score(ytest,predy),"\n")
print("classification report",classification_report(ytest,predy))
```

KNeighbors Classifier shows the best accuracy


```python
import joblib 
joblib.dump(model,"fruits_classification")
joblib.dump(scaler,"scaler")
```

```python
%%writefile fcapp.py
import streamlit as sl
import joblib
import numpy as np
fc=joblib.load("fruits_classification")
scaler=joblib.load("scaler")
sl.title("Fruits classification")
sl.subheader("You can use this to classify fruits ")
mass=sl.number_input("Mass",min_value=76,max_value=362,value=180,step=1)
width=sl.number_input("Width",min_value=5.8,max_value=9.6,value=8.0,step=0.1)
height=sl.number_input("Height",min_value=4.0,max_value=10.5,value=6.8,step=0.1)
colorscore=sl.number_input("Color Score",min_value=0.55,max_value=0.93,value=0.59,step=0.01)
if sl.button("Predict"):
  ar=np.array([mass,width,height,colorscore])
  classified=scaler.transform(ar.reshape(1,-1))
  pred=fc.predict(classified)
  sl.subheader(f"The fruit is classified as {pred[0]}")
```
